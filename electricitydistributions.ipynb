# Distribution of CFPj for Solar, Wind, Geothermal and Baseline US Energy

The distributions as described by the IPCC's Annex II, page 982: www.ipcc-wg3.de/report/IPCC_SRREN_Annex_II.pdf

From the IPCC data we observe that the distributions are positively skewed, we will use log-normal distributions for sampling.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import lognorm
```

```python
# Solar PV Electricity

min_val = 5
p25 = 29
p50 = 46
p75 = 80
max_val = 217

# Calculate mu and sigma for lognormal
mu = np.log((p50**2) / np.sqrt(p25 * p75))
sigma = np.log(p75 / p50)

print("Mu:", mu)
print("Sigma:", sigma)

# Sample from lognormal distribution
s = np.random.lognormal(mu, sigma, 10000)

# Plot histogram
count, bins, ignored = plt.hist(s, 100, density=True)

# Plot lognormal pdf
x = np.linspace(min_val, max_val, num=100)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2))
       / (x * sigma * np.sqrt(2 * np.pi)))
plt.plot(x, pdf, linewidth=2, color='r')

plt.title('Solar PV Sampled values and Log-normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```
