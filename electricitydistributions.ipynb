# Distribution of CFPj for Solar, Wind, Geothermal and Baseline US Energy

The distribution as described by the IPCC's Annex II
 Minimum:5, 25th percentile:29, 50th percentile:46, 75th percentile:80 and Maximum:217.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import lognorm
```

```python
# Solar Electricity, median, mu and sigma values
median = 46
mu = 3.83
sigma = 0.75

# Number of samples to draw
n_samples = 1000

# Draw samples from the log-normal distribution
samples = lognorm.rvs(s=sigma, scale=median, size=n_samples)

# Generate values for plotting the PDF
x = np.linspace(5, 217, 1000)
pdf_values = lognorm.pdf(x, sigma, scale=np.exp(mu))

# Plot the histogram of the sampled values alongside the PDF
plt.figure(figsize=(12, 6))

# Plot histogram of sampled values
plt.hist(samples, bins=50, density=True, alpha=0.6, label='Sampled Values (Histogram)')

# Plot the PDF
plt.plot(x, pdf_values, label=f'Log-Normal PDF\n($\\mu$={mu:.2f}, $\\sigma$={sigma:.2f})', linewidth=2)

plt.title('Sampled Values and Log-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
